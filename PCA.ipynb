```python
import pandas as pd
df= pd.read_csv('/content/Iris.csv')

df.shape
```

```python
df.head()
```

```python
df = df.drop(['Id'],axis = 1)
```

```python
df.head()
```

```python
y=df['Species']
x=df.drop(['Species'],axis=1)
x.shape
```

```python
x.head()
```

```python
import numpy as np
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
```

```python
reduced_dimention = pca.fit_transform(x)
```

```python
reduced_dimention
```

```python
print(pca.explained_variance_ratio_)
```

```python
principal_df = pd.DataFrame(data = reduced_dimention , columns = ['PC1','PC2','PC3'])
```

```python
principal_df['Species']=y
principal_df
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
 
plt.figure(figsize = (6,6))
sb.scatterplot(data = principal_df , x = 'PC1',y = 'PC2' , hue = 'Species' , s = 60 , palette= 'icefire')
```
